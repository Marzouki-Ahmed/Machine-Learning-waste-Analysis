### Cleaning Data 

```python

# Clean column names by removing any leading or trailing whitespaces.
df_GenWaste.columns = df_GenWaste.columns.str.strip()

# showing All new columns names( for checking previous operation)
df_GenWaste.columns
```

```python
# this code Convert the 'VALUE' column to a numeric data type.
# The 'errors=coerce' parameter ensures that any non-numeric values (such as text, symbols, or empty strings)
# are transformed into NaN (Not a Number) instead of raising an error and stopping the program execution(

df_GenWaste['VALUE'] = pd.to_numeric(df_GenWaste['VALUE'], errors='coerce')
df_GenWaste.head(10)
```

```python
df_GenWaste['Hazardousness'].unique() # # Displaying all unique values present in the 'Hazardousness' column
```

```python


# (Clean data in the columns that will be used for merging
#( We will manually remove identifying codes (like [HAZ_NHAZ]) using a simple "replace" function)
# we replace 'Hazardous and non-hazardous - Total[HAZ_NHAZ]' by Total
# regex=False tells Pandas to treat the pattern as a literal string, not a regular expression regex (regex is language))
# an update ( We need to to remove this code below, if we want to do that for 'Hazardousness' ..... the data still non 100% cleaned)

# We replace 'Hazardous and non-hazardous - Total[HAZ_NHAZ]' by 'Non-hazardous'
df_GenWaste['Hazardousness'] = df_GenWaste['Hazardousness'].str.replace('Hazardous and non-hazardous - Total[HAZ_NHAZ]', 'Total', regex=False)

# we replace Hazardous[NHAZ]'by 'Non-hazardous'
df_GenWaste['Hazardousness'] = df_GenWaste['Hazardousness'].str.replace('Hazardous[HAZ]', 'Hazardous', regex=False)

# we replace Non-hazardous[NHAZ]'by 'Non-hazardous'
df_GenWaste['Hazardousness'] = df_GenWaste['Hazardousness'].str.replace('Non-hazardous[NHAZ]', 'Non hazardous' , regex=False)

# Cleaning 'Waste Category' column using the same simple approach) 
# Here we just trim spaces which often cause issues during merging 
# this code Cleans spaces ='whitespace' from both ends
df_GenWaste['Waste Category'] = df_GenWaste['Waste Category'].str.strip()

# Display the first few rows of the cleaned data
df_GenWaste.tail(10)
```

```python

df_GenWaste.columns # Display the names of all columns in the DataFrame
```

```python
# Drop the 'UNIT' column and rename 'VALUE' column to 'VALUE(Tonnes)' for simplicity.
df_GenWaste = df_GenWaste.drop(columns=['UNIT']).rename(columns={'VALUE': 'VALUE(Tonnes)'})
```

```python
# Final Trim to remove any hidden spaces after replacement
# After replacing, we strip again to make sure no spaces were left behind
df_GenWaste['Waste Category'] = df_GenWaste['Waste Category'].str.strip()

# Convert VALUE to numeric (Coerce errors will handle missing data)
df_GenWaste['VALUE(Tonnes)'] = pd.to_numeric(df_GenWaste['VALUE(Tonnes)'], errors='coerce')
```
